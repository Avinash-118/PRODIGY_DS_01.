```python
import matplotlib.pyplot as plt

# Assuming 'data_df' is your DataFrame
years = data_df.columns[1:].astype(str)  # Convert years to strings
total_values = data_df[years].sum()

# Convert total_values to numeric, handling any potential NaN or non-numeric issues
total_values = total_values.apply(pd.to_numeric, errors='coerce').fillna(0)

plt.figure(figsize=(30, 30))
plt.barh(years, total_values, color='#191970')
plt.xlabel('Total Values', size=20)
plt.ylabel('Year', size=20)
plt.title('Total Values per Year', size=20)
plt.show()
```
